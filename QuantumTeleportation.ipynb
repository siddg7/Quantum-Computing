```python
from qiskit import *
qc = QuantumRegister(3)
cr0 = ClassicalRegister(1)
cr1 = ClassicalRegister(1)
circuit = QuantumCircuit(qc,cr0,cr1)
```

```python
%matplotlib inline 
```

```python
circuit.draw(output = 'mpl')
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output = 'mpl')
```

```python
circuit.h(1)
circuit.cx(1,2)
```

```python
circuit.draw(output = 'mpl')
```

```python
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output = 'mpl')
```

```python
circuit.barrier()
circuit.measure(0,0)
circuit.measure(1,1)
```

```python
circuit.draw(output = 'mpl')
```

```python
circuit.x(2).c_if(cr0, 1) 
circuit.z(2).c_if(cr1, 1)
```
